```python
import pandas as pd

#import our data_set
data_set = pd.read_csv('data_set.csv')
data_set.info()
data_set.head()
data_set.describe()
```

```python
#show the most commented topics (unique users)
from matplotlib import pyplot as plt

plt.figure(figsize=(12,6))
data_set.groupby('Title')['User'].count().plot.bar()
plt.grid()
plt.show()
```

```python
#show which users has commented which topic

plt.figure(figsize=(120,60))
data_set.groupby('User')['Title'].count().plot.bar()
plt.grid()
plt.show()
```
