# Matplot, libreria per creare grafici

Bisogna innanzitutto definire un vettore arbitrario di valori x estratti dal dominio della funzione, calcolare i rispettivi punty y tramite la formula analitica della funzione ed infine eseguire il plot dei valori tramite la libreria

```python
import numpy as np
from matplotlib import pyplot as plt
x = np.linspace(-100,100,300)
y = x**2

plt.plot(x,y)
plt.xlabel('Asse x')
plt.ylabel('Asse y')
plt.title('GRAFIC1')
plt.show()
```

# Se volessimo confrontare vari grafici possiamo utilizzare subplot

utilizzando plt.figure(figsize=16,10)) andiamo a creare una figura di dimensione 16x10 inches
successivamente possiamo andare a dividere la figura in più grafici tramite subplot come segue

```python
plt.figure(figsize=(16,10))
plt.subplot(2,2,1)
x = np.linspace(-100,100,300)
plt.plot(x**2)
plt.title('x^2')

plt.subplot(2,2,2)
x = np.linspace(1,100,300)
plt.plot(np.log(x))
plt.title('log(x)')

plt.subplot(2,2,3)
x = np.linspace(1,100,300)
plt.plot(x**2 * np.log(x))
plt.title('x*2 log(x)')

plt.subplot(2,2,4)
x = np.linspace(1,100,300)
plt.plot(np.log(x)/x)
plt.title('log(x)/x')
```

E' anche possibile modificare lo stile del grafico aggiungendo parametri alla funzione .plot() come ad esempio .plot(x,y, '--'/'.'/'o'/'r')

# Libreria pandas, le series

Una series è una struttura monodimensionale simile ad un array. Ad ogni elemento della serie viene asseggnata automaticamente un'etichetta (0 per il primo elemento, 1 per il secondoe ecc..) oppure possiamo impostarla noi passando una variabile alla funzione .Series([elementi], etichetta = [....]). 
Per velocizzare il lavoro è possibile inizializzare una series tramite un dizionario che associa ad ogni elemento la propria etichetta tramite il concetto di chiave nei dizionari ES.({'a': 1, ... }. Può essere resettata la numerazione degli indici tramite la funzione.reset_index(drop=True)
